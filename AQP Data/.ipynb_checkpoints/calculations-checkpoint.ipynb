```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('aqp-1gb.csv')
df.head(10)
```

```python
mainDBTime = df['Main DB Time (ms)']
synopsisDBTime = df['Synopsis DB Time (ms)']
mainDBTime
```

```python
mainDBMean = np.mean(mainDBTime).round(2)
mainDBMean
```

```python
synopsisDBMean = np.mean(synopsisDBTime).round(2)
synopsisDBMean
```

```python
speed = mainDBMean/synopsisDBMean
speed
```

```python
df2 = pd.read_csv('aqp-10gb.csv')
df2.head(10)
```

```python
mainDBTime2 = df2['Main DB Time (ms)']
synopsisDBTime2 = df2['Synopsis DB Time (ms)']
mainDBTime2
```

```python
mainDBMean2 = np.mean(mainDBTime2).round(2)
mainDBMean2
```

```python
synopsisDBMean2 = np.mean(synopsisDBTime2).round(2)
synopsisDBMean2
```

```python
speed2 = mainDBMean2/synopsisDBMean2
speed2
```
